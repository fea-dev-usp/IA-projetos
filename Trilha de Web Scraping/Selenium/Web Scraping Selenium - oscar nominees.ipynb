```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time
```

```python
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
driver = webdriver.Chrome()

driver.get("https://www.oscars.org/oscars/ceremonies/2024")
print(driver.title)

search = driver.find_element(By.ID, "block-quicktabs-honorees")

try:
    main = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.ID, "block-quicktabs-honorees"))
    )
    
    section = main.find_elements(By.XPATH, "//div[contains(@class, 'views-field')]")
    for i in section[5:]:
        info = i.find_elements(By.CLASS_NAME, "field-content")
        for j in info:
            print(j.text)
    
finally:
    driver.quit()
```
