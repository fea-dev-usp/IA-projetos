```python
import requests
import scrapy
from scrapy import Selector
import pandas as pd

url = "https://br.ign.com/cinema-tv"

response = requests.get(url)
html_content = response.text

class MySpider(scrapy.Spider):
    name = 'my_spider'
    start_urls = ['https://br.ign.com/cinema-tv']

    def parse(self, response):
        selector = Selector(response)
        title = selector.css('title::text').get()
        print(title)

        elements = selector.css('h3 ::text').getall()[1:25]

        scrapped_data = {
            'Títulos das notícias': elements,
        }

        df2 = pd.DataFrame(scrapped_data)
        df2.to_csv('ign.csv', index=False, encoding='utf-8')

spider = MySpider()
spider.parse(response)
```
