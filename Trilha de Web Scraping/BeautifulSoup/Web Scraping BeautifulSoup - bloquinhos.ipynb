```python
from bs4 import BeautifulSoup
import requests
import re
```

```python
import pandas as pd
url = 'https://exame.com/pop/carnaval-2024-em-sp-veja-datas-e-horarios-dos-blocos-por-regiao/'
result = requests.get(url)
doc = BeautifulSoup(result.text, 'html.parser')

news_body = doc.find(id='news-body')
uls = news_body.find_all('ul', class_='sc-1761b25b-0 kAULtx body-extra-large')
lis = news_body.find_all('li')
uls = uls[1:]
bairros = news_body.find_all('h3', class_='sc-1761b25b-0 kAULtx headline-small')

title = doc.find('h2', class_='sc-1761b25b-0 kAULtx headline-small')
print(f'{title.text}\n')

df = pd.DataFrame(columns=['Bairro', 'Bloco'])

for bairro, ul in zip(bairros, uls):
    bairro_name = bairro.text
    for li in ul.find_all('li'):
        bloco_name = li.text
        df = pd.concat([df, pd.DataFrame({'Bairro': [bairro_name], 'Bloco': [bloco_name]})], ignore_index=True)


df[['Nome do bloco', 'Horário', 'Local', 'Data']] = df['Bloco'].str.split('|', expand=True)
df.drop('Bloco', axis=1, inplace=True)

df['Horário'] = df['Horário'].str.replace('Horário:', '')
df['Local'] = df['Local'].str.replace('Local:', '')
df['Data'] = df['Data'].str.replace('Data:', '')

df
```
